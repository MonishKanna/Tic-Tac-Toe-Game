```python
import random

                                                               #creating board for Tic-Tac-Toe game
board = []
for row in range(3):
    temp = []
    for column in range(3):
        temp.append(" ")
    board.append(temp)


                                                                # function to check if the board is filled
def is_board_filled():
    for row in board:
        if " " in row:
            return False
    return True


                                                                # function to check if a player has won
def is_winner(current_player):
                                                                # check rows to find result
    for row in board:
        if row.count(current_player) == 3:
            return True
                                                                # check columns to find result
    for column in range(3):
        if board[0][column] == board[1][column] == board[2][column] == current_player:
            return True
                                                                # check diagonals to find results
    if board[0][0] == board[1][1] == board[2][2] == current_player or board[0][2] == board[1][1] == board[2][0] == current_player:
        return True
    return False

                                                                #prints board after updating user entry for reference
def print_board():
    print('   1    2    3')
    print('1  {} | {} | {}'.format(board[0][0], board[0][1], board[0][2]))
    print('  ---+---+---')
    print('2  {} | {} | {}'.format(board[1][0], board[1][1], board[1][2]))
    print('  ---+---+---')
    print('3  {} | {} | {}'.format(board[2][0], board[2][1], board[2][2]))


                                                                # function to start the game
def play_game():
                                                                # randomly select the first player
    players = ['X', 'O']
    current_player = random.choice(players)
    print('Decided {} plays first'.format(current_player))
    while True:
        print_board()
        row = int(input('Enter the row number (1-3): '))
        column = int(input('Enter the column number (1-3): '))
        if board[row-1][column-1] != " ":
            print('That spot is already taken. Try again.')
            continue
        board[row-1][column-1] = current_player
        if is_winner(current_player) == True:
            print_board()
            print('Player {} wins!'.format(current_player))
            break
        if is_board_filled() == True:
            print_board()
            print('The game is a draw!')
            break
        current_player = 'X' if current_player == 'O' else 'O'

                                                                # start the game
play_game()
```
